## Exercise

Run the necessary queries that will provide us with the following information. Compare your queries with the solutions at the end of this notebook.

### Exercise 1

Display the average unit price of products in the 'Products' table.

```python
%%sql
SELECT
    ROUND(AVG(UnitPrice),2) AS Average_Unit_Price
FROM
    Products
```

### Exercise 2


Show the total quantity of products sold for each ProductID. Use the 'OrderDetails' table.

```python
%%sql
SELECT
    ProductID,
    SUM(Quantity)
FROM 
    OrderDetails
GROUP BY
    ProductID
```

### Exercise 3


Find the maximum, minimum, and average freight from the 'Orders' table.

```python
%%sql

SELECT
    MAX(Freight),
    MIN(Freight),
    ROUND(AVG(Freight),2)
FROM
    Orders
```

### Exercise 4

How many unique customers have made orders? Use the 'Orders' table.

```python
%%sql

SELECT
    COUNT(DISTINCT CustomerID)
FROM
    Orders;
```

### Exercise 5

What is the total price for each order? You need to multiply the unit price by the quantity and subtract the discount for each order detail, then sum all the order details for each order. Use the 'OrderDetails' table.

```python
%%sql

SELECT
    OrderID,
    SUM((UnitPrice * Quantity) - Discount) AS Total_order_value
FROM 
    OrderDetails
GROUP BY
    OrderID
```

### Challenge question

What are the 5 largest orders processed? (Hint: Use the 'OrderDetails' table.)

```python
%%sql

SELECT
    OrderID,
    SUM(Quantity) AS Total_Quantity
FROM 
    OrderDetails
GROUP BY
    OrderID
ORDER BY Total_quantity DESC
```

## Solutions

### Exercise 1

```python
%%sql

SELECT 
    AVG(UnitPrice) 
FROM 
    Products;
```

### Exercise 2

```python
%%sql

SELECT 
    ProductID,
    SUM(Quantity) as TotalQuantity
FROM 
    OrderDetails
GROUP BY 
    ProductID;
```

### Exercise 3

```python
%%sql

SELECT 
    MAX(Freight) as MaxFreight,
    MIN(Freight) as MinFreight,
    AVG(Freight) as AvgFreight
FROM 
    Orders;
```

### Exercise 4

```python
%%sql

SELECT 
    COUNT(DISTINCT CustomerID) 
FROM 
    Orders;
```

### Exercise 5

```python
%%sql

SELECT 
    OrderID, 
    SUM(UnitPrice * Quantity * (1 - Discount)) as TotalPrice 
FROM 
    OrderDetails
GROUP BY 
    OrderID;
```

### Challenge question

```python
%%sql

SELECT 
    OrderID, 
    SUM(Quantity) as TotalQuantity 
FROM 
    OrderDetails
GROUP BY 
    OrderID
ORDER BY 
    TotalQuantity DESC 
```
